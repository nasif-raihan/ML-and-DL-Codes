```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.sample(2)
```

```python
titanic = sns.load_dataset('titanic')
titanic.sample(2)
```

# 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['time'], style=tips['smoker'], size=tips['sex'])
plt.title("Scatterplot (Numerical - Numerical)")
plt.plot()
```

# 2. Bar Plot (Numerical - Categorical)

```python
sns.barplot(x=titanic['pclass'], y=titanic['age'], hue=titanic['sex'])
plt.title("Bar Plot (Numerical - Categorical)")
plt.plot()
```

# 3. Box Plot (Numerical - Categorical)


```python
sns.boxplot(x=titanic['sex'], y=titanic['age'], hue=titanic['survived'])
plt.title("Box Plot (Numerical - Categorical)")
plt.plot()
```

# 4. kdeplot (Numerical - Categorical)

```python
sns.kdeplot(data=titanic[titanic['survived'] == 0]['age'], label='Not Survived', fill=True)
sns.kdeplot(data=titanic[titanic['survived'] == 1]['age'], label='Survived', fill=True)
plt.title("kdeplot (Numerical - Categorical)")
plt.plot()
```

# 5. HeatMap (Categorical - Categorical)

```python
df = pd.crosstab(titanic['embarked'],titanic['survived'])
df.head()
```

```python
sns.heatmap(data=df, annot=True, fmt="d")
plt.title("HeatMap (Categorical - Categorical)")
plt.plot()
```

# 6. ClusterMap (Categorical - Categorical)

```python
# df = pd.crosstab(titanic['parch'],titanic['survived'])
df = pd.crosstab(tips['size'], tips['time'])
df.head()
```

```python
sns.clustermap(df)
plt.title("ClusterMap (Categorical - Categorical)")
plt.plot()
```

# 7. Pairplot

```python
iris = sns.load_dataset('iris')
iris.sample(2)
```

```python
sns.pairplot(data=iris, hue='species')
plt.title("PairPlot")
plt.plot()
```

# 8. Lineplot (Numerical - Numerical)

```python
flights = sns.load_dataset('flights')
flights.sample(2)
```

```python
df = flights.groupby('year')['passengers'].sum().reset_index()
df.head()
```

```python
sns.lineplot(x=df['year'], y=df['passengers'])
plt.title("LinePlot")
plt.plot()
```

```python
sns.lineplot(x=flights['month'], y=flights['passengers'], hue=flights['year'])
plt.title("LinePlot")
plt.plot()
```

```python
from skimage import data
from skimage import color
from skimage.filters import try_all_threshold
from skimage.filters import threshold_otsu
from skimage.filters import threshold_local
from PIL import Image
from pylab import *
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np
```

```python
drive.mount("/content/data/")
```

```python
path = '/content/data/MyDrive/Colab Notebooks/Datasets/brazil.jpeg'

img = array(Image.open(path))

imshow(img)
```

```python
g_img = color.rgb2gray(img)

imshow(g_img)
```

```python
rgb_img = color.gray2rgb(g_img)

imshow(rgb_img)

gray_img = color.rgb2gray(rgb_img)

#imshow(gray_img)
```

```python
def show_image(image, title='Image', cmap_type='gray'):
  plt.imshow(image, cmap = cmap_type)
  plt.title(title)
  plt.axis('off')
  plt.show()

#show_image(rgb_img)
```

```python
red = img[:, :, 0]
green = img[:, :, 1]
blue = img[:, :, 2]

show_image(img, "Original")
show_image(red, "RED")
show_image(green, "GREEN")
show_image(blue, "BLUE")
```

```python
print("Shape: ", img.shape)
print("Size: ", img.size)
```

```python
v_img = np.flipud(img)

show_image(v_img, "Vertically Flipped")
```

```python
h_img = np.fliplr(img)

show_image(h_img, "Horizontally Flipped")
```

```python
plt.hist(red.ravel(), bins=256)
plt.title("Red Histogram")
plt.show()
```

```python
thresh = threshold_otsu(gray_img)

inverted_binary = gray_img < thresh ## why always black??

show_image(gray_img)
show_image(inverted_binary, "Inverted Threshold")
fix, ax = try_all_threshold(gray_img, verbose=False)
```

```python
block_size = 35

local_thresh = threshold_local(gray_img, block_size, offset=10)
binary_local = gray_img > local_thresh
show_image(binary_local)
```

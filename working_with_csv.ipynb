## parse_dates parameter

````python
df = pd.read_csv(ipl_matches_url)
df['date'].info()
````

````python
df = pd.read_csv(ipl_matches_url, parse_dates=['date'])
df['date'].info()
````

## converters parameter

````python
def rename(name):
    words = name.split( )
    if len(words) > 2:
        return f"{words[0][0]}{words[1][0]}{words[2][0]}"
    return name
````

````python
df = pd.read_csv(ipl_matches_url, converters={'team1':rename, 'team2':rename})
df.head()
````

## na_values parameter

Specifying some text to be considered as as NaN value

````python
df = pd.read_csv(aug_train_url)
print(df.shape)
````

````python
df['gender'].value_counts()
````

````python
df['gender'].unique()
````

````python
df = pd.read_csv(aug_train_url, na_values=['Other',])
df['gender'].value_counts()
````
